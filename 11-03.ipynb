```r
x = (0:100)/10
y = 1 + x + 2 * sin(x) + rnorm(length(x))
yave = 1 + x + 2*sin(x)
```

```r
plot(1:101, y)
lines(1:101, yave)
```

```r
model <- lm(y~x + sin(x))
```

```r
summary(model)
```

```r
png('test.png')
plot(1:101, y)
lines(1:101, yave)
lines(1:101, model$fitted.values, col='red')
dev.off()
```

```r
pred=predict(model, level=.8, i='c')
pred1=predict(model, level=.85, i='p')
```

```r
x = (0:100)/10
y = 1 + x + 2 * sin(x) + runif(length(x), -1, 1)
yave = 1 + x + 2*sin(x)
```

```r
model <- lm(y~x + sin(x))
```

```r
summary(model)
```

```r
pred=predict(model, level=.8, i='c')
pred1=predict(model, level=.85, i='p')
```

```r
plot(1:101, y)
lines(1:101, model$fitted.values, col='green')
lines(1:101, pred[,'lwr'], col='blue')
lines(1:101, pred[,'upr'], col='red')
lines(1:101, pred1[,'lwr'], col='gray')
lines(1:101, pred1[,'upr'], col='gray')
```

```r
plot(model$fitted.values, model$res)
```
